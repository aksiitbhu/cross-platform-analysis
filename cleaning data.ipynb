```python
import pandas as pd
import re
from ast import literal_eval
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
data=pd.read_csv('/content/drive/MyDrive/Discord-Reddit/Datasets/dirty data/Ukrainian_conflict1.csv',sep='delimiter',header=None)
```

```python
def str_dict_to_csv_col(row):
  row = re.sub(',,+','',row)
  row = re.sub('""+','"',row)
  row = re.sub('","',",",row)
  row = re.sub(',"',",",row)
  row = re.sub( '",(?= [^\'])' , ",", row)
  
  row = re.sub( '\'content\': (?=[a-zA-Z0-9])' ,'\'content\': "' , row)
  row = re.sub(", \'message_reference\':.*","}",row)
  row = re.sub("\'attachments\':.* \'timestamp\':","\'timestamp\':",row)

  # row = re.sub("(?<='content': \")'(?=\" , 'channel_id':)", "", row)
  row = re.sub('\'"',"'",row)

  # return row
  try:
    row = literal_eval(row)

    return pd.Series([row['id'],row['type'],row['content'],row['channel_id'],
                      row['author']['id'],row['author']['username'],
                      row['timestamp']])
  except:
    print('error in a line')
    return pd.Series(['error','error','error','error',
                      'error','error',
                      'error'])
```

```python
data_parsed = pd.DataFrame()
```

```python
data_parsed[['id','type','content','channel_id','author_id','author_username','timestamp']] = data[0].progress_apply(str_dict_to_csv_col)
```
